```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_cleaned = df.dropna(subset=["Total", "Year","Cement","Flaring","Other"])
data_cleaned
```

```python
data_cleaned["Year"] = pd.to_numeric(data_cleaned["Year"])
```

```python
data_filtered = data_cleaned[data_cleaned["Total"] > 0]
```

```python
# 1. Stacked Area Chart for Emission Sources Over Time (Global)
emission_sources = ["Coal", "Oil", "Gas"]
stacked_data = data_filtered.groupby("Year")[emission_sources].sum()
```

```python
plt.figure(figsize=(12, 6))
stacked_data.plot(kind="area", stacked=True, figsize=(12, 6))
plt.title("Global Emissions by Source Over Time")
plt.xlabel("Year")
plt.ylabel("Emissions (Mt)")
plt.legend(title="Emission Sources")
plt.tight_layout()
plt.show()
```

```python
# 2. Geospatial Heatmap of Emissions Per Capita by Country (Latest Year with Data)
latest_year = data_filtered["Year"].max()
latest_year_data = data_filtered[data_filtered["Year"] == latest_year]
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    x="Per Capita",
    y="Country",
    data=latest_year_data.nlargest(15, "Per Capita").sort_values("Per Capita", ascending=True),
    palette="viridis",
)
plt.title(f"Top 15 Countries by Per Capita Emissions in {latest_year}")
plt.xlabel("Per Capita Emissions (Mt)")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
# 3. Bar Chart Comparing Total Emissions by Country (Latest Year with Data)
plt.figure(figsize=(12, 6))
sns.barplot(
    x="Total",
    y="Country",
    data=latest_year_data.nlargest(15, "Total").sort_values("Total", ascending=True),
    palette="coolwarm",
)
plt.title(f"Top 15 Countries by Total Emissions in {latest_year}")
plt.xlabel("Total Emissions (Mt)")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```
