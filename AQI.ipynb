```python
!pip install pycountry
import pycountry #importing pycountry

# Function to get the three-letter ISO alpha-3 code for a country name
def get_country_code(country_name):
    try:
        return pycountry.countries.lookup(country_name).alpha_3
    except LookupError:
        return None  # If the country is not found

# Load the DataFrame and assign it to global_ap_df
global_ap_df = pd.read_csv(r'global AP dataset.csv')

# Apply the function to the column containing country names
# Replace 'Country' with the actual column name in your dataset
global_ap_df['Country_Code'] = global_ap_df['Country'].apply(get_country_code)

# Display the first few rows to check the results
global_ap_df[['Country', 'Country_Code']]
```

```python
visual = global_ap_df.select_dtypes(include='int64')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(visual['AQI Value'], kde=True, bins=30, color='blue')
plt.title('Distribution of Overall AQI Values', fontsize=16)
plt.xlabel('AQI Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(visual, orient="h", palette="Set2")
plt.title('Boxplot of AQI Metrics', fontsize=16)
plt.xlabel('AQI Value', fontsize=12)
plt.ylabel('Metrics', fontsize=12)
plt.show()


```

```python
# 2. Correlation heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = visual.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```
